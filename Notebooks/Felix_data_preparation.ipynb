#### PATIENT DEMOGRAPHICS

```python
import pandas as pd
from datetime import datetime
import numpy as np


dfs = []
for year in ['2008', '2009', '2010']:
    df = pd.read_csv(os.path.join(root_dir,f'DE1_0_{year}_Beneficiary_Summary_File_Sample_20.csv'))
    df['Year'] = year
    dfs.append(df)


ben_df = pd.concat(dfs, ignore_index=True)


ben_sets = {year: set(df['DESYNPUF_ID']) for year, df in ben_df.groupby('Year')}


ben_df['NewEnrollee'] = False
for year, ben_set in ben_sets.items():
    if year == '2008': # no previous year to check
        continue
    prev_year = str(int(year) - 1)
    ben_df.loc[(ben_df['Year'] == year) & (~ben_df['DESYNPUF_ID'].isin(ben_sets[prev_year])), 'NewEnrollee'] = True


ben_df.rename(columns={'DESYNPUF_ID':'MBI',
                       'BENE_SEX_IDENT_CD':'SEX',
                       'BENE_BIRTH_DT':'DOB'}, inplace=True)


ben_df['LTIMCAID'] = np.where(ben_df['MEDREIMB_IP'] > 0, 1, 0)


ben_df['NEMCAID'] = np.where(ben_df['NewEnrollee'] & (ben_df['MEDREIMB_IP'] > 0), 1, 0)


ben_df['DOB'] = pd.to_datetime(ben_df['DOB'], format='%Y%m%d')
ben_df['AGE'] = ben_df['Year'].astype(int) - ben_df['DOB'].dt.year


conditions = [
    (ben_df['AGE'] >= 65), # old age
    (ben_df['AGE'] < 65) & (ben_df['BENE_ESRD_IND'] == 0), # disability
    (ben_df['BENE_ESRD_IND'] == 1), # ESRD
]
choices = ['0', '1', '2'] # corresponding OREC codes

ben_df['OREC'] = np.select(conditions, choices, default='0') # default to old age if none of conditions met


ben_df = ben_df[['MBI', 'SEX', 'DOB', 'LTIMCAID', 'NEMCAID', 'OREC']]


ben_df.to_csv('person_file.csv', index=False)
ben_df.head()
```

```python
import pandas as pd


df_carrier_a = pd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Carrier_Claims_Sample_20A.csv'))


df_carrier_a_claim_level = df_carrier_a.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], 
                                             value_vars=[f'ICD9_DGNS_CD_{i}' for i in range(1, 9)], 
                                             value_name='DIAG')


df_carrier_a_line_level = df_carrier_a.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], 
                                            value_vars=[f'LINE_ICD9_DGNS_CD_{i}' for i in range(1, 14)], 
                                            value_name='DIAG')


df_carrier_a = pd.concat([df_carrier_a_claim_level, df_carrier_a_line_level])


df_carrier_b = pd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Carrier_Claims_Sample_20B.csv'))


df_carrier_b_claim_level = df_carrier_b.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], 
                                             value_vars=[f'ICD9_DGNS_CD_{i}' for i in range(1, 9)], 
                                             value_name='DIAG')


df_carrier_b_line_level = df_carrier_b.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], 
                                            value_vars=[f'LINE_ICD9_DGNS_CD_{i}' for i in range(1, 14)], 
                                            value_name='DIAG')


df_carrier_b = pd.concat([df_carrier_b_claim_level, df_carrier_b_line_level])


df_carrier = pd.concat([df_carrier_a, df_carrier_b])


df_carrier = df_carrier.dropna(subset=['DIAG'])


df_carrier.to_csv('DIAG.csv', index=False)
df_carrier.head()
```

```python
import dask.dataframe as dd


ddf_carrier_a = dd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Carrier_Claims_Sample_20A.csv'), dtype=str)
ddf_carrier_b = dd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Carrier_Claims_Sample_20B.csv'), dtype=str)


def reshape_ddf(ddf):
    ddf_claim_level = ddf.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'],
                               value_vars=[f'ICD9_DGNS_CD_{i}' for i in range(1, 9)],
                               value_name='DIAG')
    ddf_line_level = ddf.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'],
                              value_vars=[f'LINE_ICD9_DGNS_CD_{i}' for i in range(1, 14)],
                              value_name='DIAG')
    return dd.concat([ddf_claim_level, ddf_line_level])


ddf_carrier_a = reshape_ddf(ddf_carrier_a)
ddf_carrier_b = reshape_ddf(ddf_carrier_b)


ddf_carrier = dd.concat([ddf_carrier_a, ddf_carrier_b])


ddf_carrier = ddf_carrier.dropna(subset=['DIAG'])

ddf_inpatient = dd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Inpatient_Claims_Sample_20.csv'), dtype=str, assume_missing=True)


ddf_outpatient = dd.read_csv(os.path.join(root_dir,'DE1_0_2008_to_2010_Outpatient_Claims_Sample_20.csv'), dtype=str, assume_missing=True)

ddf_inpatient = ddf_inpatient.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], value_vars=[f'ICD9_DGNS_CD_{i}' for i in range(1, 9)], value_name='DIAG')
ddf_outpatient = ddf_outpatient.melt(id_vars=['DESYNPUF_ID', 'CLM_FROM_DT'], value_vars=[f'ICD9_DGNS_CD_{i}' for i in range(1, 9)], value_name='DIAG')


ddf_all = dd.concat([ddf_carrier, ddf_inpatient, ddf_outpatient])


ddf_all.head()    
```

```python
import dask
with dask.config.set({
    'dataframe.output.max_rows': 300,   
    'dataframe.output.max_columns': 50  
}):
    display(ddf_all.compute())
```
